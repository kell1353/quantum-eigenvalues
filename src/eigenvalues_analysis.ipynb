# Analyzing results

We will use python and this jupyter notebook to plot and analyze the results obtained from the
fortran source code.

## Plotting results

For this we will use the `numpy` and `matplotlib.pyplot` packages

```python
# Importing numpy and matplotlib
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# different styles are predefined that give different appearance to the plots. This one emulates the style of 
# the figures you'll find on statisical analysis news site fivethirtyeight.com
plt.style.use('fivethirtyeight')
```

```python
# reading data stored in infinite_well_wf.dat

file_name = 'infinite_well_wf.dat'

# loadtxt reads all the data in a file and stores them in an array (assuming it's all floats)
# we use skiprows=1 to avoid the header
all_data = np.loadtxt(file_name,skiprows=1)

# We transpose the array in order to be able to extract each column individually
all_data = all_data.transpose()

well_x_sample_points = all_data[0]
well_ground_state = all_data[1]
well_1st_excited = all_data[2]
well_2nd_excited = all_data[3]
```

```python
# reading data stored in harmonic_oscillator_wf.dat

file_name = 'harmonic_oscillator_wf.dat'

# loadtxt reads all the data in a file and stores them in an array (assuming it's all floats)
# we use skiprows=1 to avoid the header
all_data = np.loadtxt(file_name,skiprows=1)

# We transpose the array in order to be able to extract each column individually
all_data = all_data.transpose()

harmonic_x_sample_points = all_data[0]
harmonic_ground_state = all_data[1]
harmonic_1st_excited = all_data[2]
harmonic_2nd_excited = all_data[3]
```

```python
# reading data stored in woods_saxon_wf.dat

file_name = 'woods_saxon_wf.dat'

# loadtxt reads all the data in a file and stores them in an array (assuming it's all floats)
# we use skiprows=1 to avoid the header
all_data = np.loadtxt(file_name,skiprows=1)

# We transpose the array in order to be able to extract each column individually
all_data = all_data.transpose()

woods_saxon_x_sample_points = all_data[0]
woods_saxon_ground_state = all_data[1]
woods_saxon_1st_excited = all_data[2]
woods_saxon_2nd_excited = all_data[3]
```

### Ground State Normalized Probability Densities

#### As a function of the number of the sample points

```python
#
plt.plot(well_x_sample_points, well_ground_state, label='Infinite Well')
plt.plot(harmonic_x_sample_points, harmonic_ground_state, label='Harmonic Oscillator')
plt.plot(woods_saxon_x_sample_points, woods_saxon_1st_excited, label='Woods_Saxon')

plt.xlabel('Sample x points (L)')
plt.ylabel('Probability Density (1/L)')
# legend displays the labels given in each plot
plt.legend()
plt.show()
```

### 1st Excited State Normalized Probability Densities

#### As a function of the number of the sample points

```python
#
plt.plot(well_x_sample_points, well_1st_excited, label='Infinite Well')
plt.plot(harmonic_x_sample_points, harmonic_1st_excited, label='Harmonic Oscillator')
plt.plot(woods_saxon_x_sample_points, woods_saxon_1st_excited, label='Woods_Saxon')

plt.xlabel('Sample x points (L)')
plt.ylabel('Probability Density (1/L)')
# legend displays the labels given in each plot
plt.legend()
plt.show()
```

### 2nd Excited State Normalized Probability Densities

#### As a function of the number of the sample points

```python
#
plt.plot(well_x_sample_points, well_2nd_excited, label='Infinite Well')
plt.plot(harmonic_x_sample_points, harmonic_2nd_excited, label='Harmonic Oscillator')
plt.plot(woods_saxon_x_sample_points, woods_saxon_2nd_excited, label='Woods_Saxon')

plt.xlabel('Sample x points (L)')
plt.ylabel('Probability Density (1/L)')
# legend displays the labels given in each plot
plt.legend()
plt.show()
```

### The three lowest energies for the Woods-Saxon potential as a function of the radius 

```python
# reading data stored in woods_saxon_energy.dat

file_name = 'woods_saxon_energy.dat'

# loadtxt reads all the data in a file and stores them in an array (assuming it's all floats)
# we use skiprows=1 to avoid the header
all_data = np.loadtxt(file_name,skiprows=1)

# We transpose the array in order to be able to extract each column individually
all_data = all_data.transpose()

woods_saxon_radii = all_data[0]
woods_saxon_ground_state = all_data[1]
woods_saxon_1st_excited = all_data[2]
woods_saxon_2nd_excited = all_data[3]
```

```python
#
plt.plot(woods_saxon_radii, woods_saxon_ground_state, label='Ground State')
plt.plot(woods_saxon_radii, woods_saxon_1st_excited, label='1st Excited State')
plt.plot(woods_saxon_radii, woods_saxon_2nd_excited, label='2nd Excited State')

plt.xlabel('Radius (fm)')
plt.ylabel('Energies (MeV)')
# legend displays the labels given in each plot
plt.legend()
plt.show()
```
